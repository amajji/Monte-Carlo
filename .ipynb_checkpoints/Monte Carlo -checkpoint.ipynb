### Bienaymé-Tchebychev inequality

<div class="alert alert-danger"><b>Theorem </b> 
    
Let X be a random variable with an expectation value E(X) and variance Var(X). For any strictly postive real a : 


\begin{align}
P(|X - E(X)| ) \geq a)  \leq  \frac{Var(X)}{a^2}
\end{align}
    
</div>

### The law of large numbers

<div class="alert alert-danger"><b>Theorem </b>

We consider a sequence of random variables ($X_1$, $X_2$, ... ,$X_N$), two to two independents and following the same probability distribution (with a mean m and standard deviation σ :
    
\begin{align}
    E(X_i) = m  \;\; and \;\;  E(X_i - m)^2 = σ   \;\;\;\;\;\;   \forall i \in [1, 2, ... ,n] 
\end{align} 
    
The arithmetic mean of the n first random variables tends, in probability, towards the stochastic mean m when (n→∞) : 
   
$$ \forall h \;\;\;\; lim_{n→∞} \;\; Pr\{|\frac{X_1 + X_2 + ... + X_N}{N} - m | > h\} = 0 $$
    
</div>

Proof of theorem: Let's consider $X_0, ..., X_N $ N random variables and  $S = X_1 + X_2 + ... + X_N $ 

We know that 
                 
\begin{align}
E(S) &= E( \sum_{i = 1}^N X_{i})  \\
&= \sum_{i = 1}^N E(X_{i})  \\
&= Nm  \\
\end{align}     

and 

\begin{align}
E(S - E(S))^2 &= E( \sum_{i = 1}^N X_{i} - m )^2  \\
\end{align}

And we know that $ E{(Xi−m)(Xj−m)}=E(Xi−m)E(Xj−m)=0×0=0 $ since ($ X_1$ - m) and ($ X_j $ - m) are independents. So : 

\begin{align}
E(S - E(S))^2 &= E( \sum_{i = 1}^N X_{i} - m )^2  \\
&= \sum_{i = 1}^N (E(X_{i} - m)^2  \\
&= Nσ^2
\end{align}

So 

\begin{align}
E(\frac{S}{N} - m ) =  σ^2 \\
\end{align}

By applying the Bienaymé-Tchebychev inequality on the variable $\frac{S}{N} - m $ : 

\begin{align}
 Pr\{|\frac{S}{N} - m | > h\} \leq \frac{σ^2}{N} \frac{1}{h^2} \\
\end{align}

\begin{align}
 \forall h \;\;\;\; lim_{n→∞} \;\; Pr\{|\frac{X_1 + X_2 + ... + X_N}{N} - m | > h\} = 0
\end{align}

### Central limit theorem 

<div class="alert alert-danger"><b> Central limit theorem </b> 
Let X1, X2... be a set of random variables defined on the same probability space, following the same law D and independent. Suppose that the expectation μ and the standard deviation σ of D exist and are finite ($ \sigma \neq 0 $ ).

Consider the sum $ S_n = X_1 + ... + X_N $ . Then the expectation of $ S_n $ is nμ and its standard deviation is  $ σ N^½ $ . Moreover, to speak informally, the law of $ S_n $ tends to the normal law $ N(Nμ,Nσ^2) $ as n tends to infinity.

Let $Z_{n}$  be : 
    
\begin{align}
     Z_{n}  =  \frac{S_{n} - nμ}{Nσ} 
\end{align}

Then the law of $ Z_{n} $ converges to the normal law N(0,1) when n tends to infinity


# Monte-Carlo Methods

One of the procedures for calculating a quantity by the Monte-Carlo method is to first put it in the form of an expectation value E(X), at the end of this step, it remains to calculate this quantity by an expectation E (X) of the random variable X. For this calculation, it is necessary to know how to simulate a random variable according to the law of X. We then have a sequence $ (X_i) $ for $ \;\; 1≤ i ≤ N $ of N realizations of the random variable X . We then approximate E(X) by:


\begin{align}
      E(X) \approx \frac{X_1 + X_2 + ... + X_N}{N} 
\end{align}

## Calculation of the integral by expectation

<div class="alert alert-danger"><b>Monte Carlo Methods  </b> 


The calculation of the integral $\int_{\Omega}^{} f_{X}(x)dx $ by the method of Monte Carlo is reduced to solving the following integral:  $\int_{[0,1]^d}^{} f_{X}(x)dx $

In this case, the Monte Carlo method consists in writing this integral in the form of the expectation of f which is the generalization of the mean of f on $[0, 1]^d $, and with u a random variable following the uniform law on $[ 0, 1]^d$:

\begin{align}
      I = E(f(u))
\end{align}

We approximate the integral as follows : 

\begin{align}
      I \approx I_N = \frac{1}{N} \sum_{i = 1}^N f(x_{i})
\end{align}

We chose the values $ x_i $ from the interval $ \Omega $. The approximation is more precise as soon as the number N increases : 


\begin{align}
     \int_{\Omega}^{} f_{X}(x)dx = lim_{N→∞} I_N = lim_{N→∞} \frac{1}{N} \sum_{i = 1}^N f(x_{i})
\end{align}


### General Case : 


Suppose there is an expression for the mathematical expectation of a function g of the random variable X, resulting from the transfer theorem, according to which



\begin{align}
     I = E(g(X)) = \int_{a}^{b} g(x) f_{X}(x)dx
\end{align}

where $ f_X $ is a density function of X on the support [a, b]. It is common to take a uniform distribution over [a, b] : 

\begin{align}
     f_{X} = \frac{1}{b - a}
\end{align}

We want to approximate this integral using Monte Carlo method : 

\begin{align}
     \int_{a}^{b} g(x) f_{X}(x)dx
\end{align}

The idea is to produce a sample $(x_1 ,x_2,...,x_N)$ of the random variable X (thus according to the density $f_X$ if X is continuous or according to the law of X if X is discrete) on the support [a , b], and to calculate a so-called Monte-Carlo estimator of G, from this sample.

The law of large numbers suggests constructing this estimator from the empirical mean:


\begin{align}
     I_N = \frac{1}{N} \sum_{i = 1}^N g(x_i)
\end{align}

Based on the same theorem : 

\begin{align}
      \frac{g(x_1) + g(x_2) + ... + g(x_N)}{N} →m  
\end{align}

So the approximatio of $I$ is :
    
\begin{align}
      I \approx I_N = \frac{1}{N} \sum_{i = 1}^N g(x_{i})
\end{align}

The difference between this ratio and its limit has the standard deviation : $ \sigma / \sqrt{N} $

</div>

<div class="alert alert-info"> <b> Remark </b> : 
    The law of large numbers justifies the convergence of the method, and the central limit theorem specifies the speed of convergence.
</div>

### Applications

#### 1 - Integral calculation

We want to approximate the following integral : 

\begin{align}
     I  = \int_{0}^{2} e^{-x^2}dx
\end{align}



We must first write $I$ as a mathematical expectation by choosing a probability density f on [0, 2] that is easy to simulate. We choose for this example the density of the continuous uniform distribution on [0, 2] : $f(x) = \frac{1}{2} I_{[0,2]} $ 

The new expression become : 

\begin{align}
     I  = \int_{0}^{2} 2e^{-x^2} \frac{1}{2} dx
\end{align}

Using the Monte Carlo Method :

\begin{align}
      I = E(2e^{-X^2})  \;\;\;\; where  \;\;\;\;  X \sim U(0, 2)  
\end{align}

So : 

\begin{align}
      I_n = \frac{1}{n} \sum_{i = 1}^n  2e^{-x_i^2} 
\end{align}


Below are the steps to follow :

<div class="alert alert-info"> <b>  </b>  
1 - Generate independent values $ U_1, U_2, · · · , U_n ∼ U(0, 1)  $ 
    
2 - Computing  $ X_1, X_2, ... , X_n $  using the formula  $ X_i = 2U_i  $
    
3 - Compute $ g(X_i) = 2e^{-x_i^2} $

4 - Calculate the arithmetic mean to find $ I_n  $
</div>


#### 2 - Area calculation (Approximation of π)

we will find the approximation of π by estimating the area of a quarter disk with radius 1. 

<img src="Images/monte carlo.png" width="400px"></img>

- The area method

Consider a point M with coordinates (x, y), where 0<x<1 and 0<y<1. We randomly draw the values of x and y between 0 and 1 according to a uniform law. The point M belongs to the disk with center (0,0) of radius R=1 if and only if $x^2+y^2≤1$ and with a probability $\frac{π}{4} = \frac{σ}{S} \approx  \frac{k}{N}$ (the containing square has area $ S=R^2=1 $ ) . So the approximaion of π is $\frac{4k}{N} $ with k is the number the points drawn randomly from the circle

- The average value method

We know that the equation of a circle is $f(x) = \sqrt{1 – x^2}$ and the area under the curve is equal to : 



\begin{align}
     I  = \int_{0}^{1} \sqrt{1 – x^2} dx =  \frac{π}{4}
\end{align}

To compute the integral, we use Monte Carlo method : 
        
\begin{align}
      I \approx I_N = \frac{1}{N} \sum_{i = 1}^N g(x_{i}) = \frac{1}{N} \sum_{i = 1}^N \sqrt{1 – x_i^2}) = \frac{π}{4}
\end{align}

So 

\begin{align}
      π \approx 4 * \frac{1}{N} \sum_{i = 1}^N \sqrt{1 – x_i^2}) 
\end{align}

```python
# Importing Packages
import matplotlib.pyplot as plt
import numpy as np 
import math
import random
plt.rcParams["figure.figsize"] = (9,9)
```

```python
# number of samples
N = 100000

# Monte Carlo estimation
MC_estimation  = 0 

# Compute the monte Carlo estimation
for i in range(N) : 
    # generate a random variable X ~ U(0,1)
    xi = random.uniform(0, 1)
    
    # Compute the estimation
    MC_estimation += math.sqrt(1 - xi**2)
    
MC_estimation = (1/N)*MC_estimation 
 
print("The estimation of pi using the Monte Carlo method is : ", 4*MC_estimation)
```

#### 3 - The Dice Game

Our simple game will involve two six-sided dice. In order to win, the player needs to roll the same number on both dice. A six-sided die has 6 outcomes (1, 2, 3, 4, 5 and 6) means the player has 6/36 chance to win and the house 30/36. As we can see, in this game, the house has more opportunities to win.
Let's say our player starts with a balance of 1000 $ \$ $ and is prepared to lose it all. We assume also that the house is generous: it offers 4 $ \$ $ each time the player wins. 

```python
# function to roll dice 
def rool_dice():
    # roll the first die
    roll1 = random.randint(1,6)
    # roll the second die
    roll2 = random.randint(1,6)
    
    # Result of rolling, it takes True if the dice have the same side and False otherwise
    result = False
    
    if roll1 == roll2 : 
        result = True

    return result
```

```python
# Number of simulations 
N = 10000

# Number maximum of bets  
N_bets = 1000

# win probability 
win_probability = []
end_balance = []
```

```python
# For loop to run for the number of simulations desired
for i in range(N) :
    
    # balance value : 1000$
    balance = [1000]

    # number of rolls
    num_rolls = [0]
    
    # Rolling dice 1000 times 
    for j in range(N_bets):
        
        # roll dice 
        result_roll = rool_dice()
        if result_roll : 
            
            # Add 4$ to the balance
            balance.append(balance[-1] + 4)
            # add 1 when the player win
            win_probability.append(1)
            
        else : 
            # subtract 1$ when the player loses 
            balance.append(balance[-1] - 1)
            # add 0 when the player loses
            win_probability.append(0)
        
        # add number of iteration
        num_rolls.append(j)
        # add end balance value
        end_balance.append(balance[-1])
    
    # plot figure
    plt.plot(num_rolls, balance)
    
## Showing the plot after the simulations are finished
plt.show()
```

```python
# Averaging win probability and end balance
overall_win_probability = sum(win_probability)/len(win_probability)
overall_end_balance = sum(end_balance)/len(end_balance)
# Displaying the averages
print("Average win probability after " + str(N) + " runs: " + str(overall_win_probability))
print("Average ending balance after " + str(N) + " runs: $" + str(overall_end_balance))
```

- Conclusion 

As we can see, after 10 000 simulations, the player rarely earn more than 1 000 $ \$ $. In fact, the average of ending balance is 917 $ \$ $ , means the house has more chance to come out on top even thought it pays out 4 $ \$ $  when the player won.

Also noteded that the probability of win is 30/36 = 1/6 which  matched our Monte Carlo prediction.
